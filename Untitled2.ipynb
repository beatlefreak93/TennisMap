```python
import pandas as pd
import numpy as np

df_21 = pd.read_csv('21data.csv')
df_22 = pd.read_csv('22data.csv')
```

```python
print(df_21[df_21['상권업종소분류명'].str.contains('테니스장')].shape)
print(df_22[df_22['상권업종소분류명'].str.contains('테니스장')].shape)
```

```python
df_21['year'] = 2021
df_22['year'] = 2022
```

```python
df_t = pd.concat([df_21[df_21['상권업종소분류명'].str.contains('테니스장')],df_22[df_22['상권업종소분류명'].str.contains('테니스장')]])
```

```python
df_t.loc[df_t['year'] == 2022, '브랜드명'] = '22테니스'
```

```python
df_t.loc[df_t['year'] == 2021, '브랜드명'] = '21테니스'
```

```python
df_t['브랜드명'].value_counts()
```

```python
import folium
```

```python
map = folium.Map(location=[df_t['위도'].mean(), df_t['경도'].mean()], zoom_start=12, tiles='Stamen Toner')

for n in df_t.index:
    popup_name = df_t.loc[n, '브랜드명'] + ' - ' + df_t.loc[n, '도로명주소']
    
    if df_t['브랜드명'][n] == '22테니스':
        icon_color = 'red'
        radius = 8
    else:
            icon_color = 'green'
            radius = 18
            
    folium.CircleMarker([df_t.loc[n, '위도'], df_t.loc[n, '경도']],
                 popup=popup_name,
                 radius=radius,
                 fill=True,
                 color=icon_color).add_to(map)

map
```
